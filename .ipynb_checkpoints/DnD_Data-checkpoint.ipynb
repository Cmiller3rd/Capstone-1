In this survey of D&D character data (from Kaggle.com, link provided in Jupyter Notebook), I am starting with a couple of hypotheses. First, I assume that the most common race will be Human, and the most common class will be Fighter. I also believe that the highest stat over all characters will be Dexterity, and that the most common casting stat (ability score that determines spellcasting ability) will be Wisdom.

As a D&D player, Dungeon Master, and overall fan, I have quite of bit of prior knowledge to bring to this analysis. Depending on what I find, I may have some ideas on how to recommend a clearer balance for the game (if, for example, Dexterity is overwhelmingly preferred over other ability score stats) or how to shore up potential "dry areas" (aspects of the game that show little activity).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
from itertools import groupby

# Kaggle.com Data link: https://www.kaggle.com/datasets/joebeachcapital/dungeons-and-dragons-characters
```

```python
characters = pd.read_csv('/Users/cmiller3rd/Documents/Coding_Temple/Capstone_I/dnd_chars_unique.tsv', sep='\t')

# add character tier to the data

def char_tiers(level):
    if level >= 1 and level < 5:
        return 'Tier 1'
    elif level >= 5 and level < 11:
        return 'Tier 2'
    elif level >= 11 and level < 17:
        return 'Tier 3'
    elif level >= 17:
        return 'Tier 4'
    else:
        return 'NaN'
characters["tier"] = characters['level'].apply(char_tiers)
characters[['alias', 'tier']].groupby('tier').count().sort_values('tier')
```

```python
# drop Ability Scores over 30
filtered_Str = characters.query("(Str <= 30) & (Str >= 3)")
filtered_Con = filtered_Str.query("(Con <= 30) & (Con >= 3)")
filtered_Wis = filtered_Con.query("(Wis <= 30) & (Wis >= 3)")
filtered_HP = filtered_Wis.query("(HP > 0)")
filtered_Int = filtered_HP.query("(Int <= 30) & (Int >= 3)")
filtered_Cha = filtered_Int.query("(Cha <= 30) & (Cha >= 3)")
```

```python
filtered_chars = filtered_Cha[['alias', 'race', 'processedRace', 
                               'justClass', 'subclass', 'level', 'AC', 'Str',
                               'Dex', 'Con', 'Int', 'Wis', 'Cha', 
                               'castingStat', 'choices', 'tier']]
```

```python
avg_stat = filtered_chars.describe()
```

```python
popular_class = filtered_chars[['justClass', 'alias']].groupby('justClass').count().sort_values('alias', ascending=False)
```

```python
import regex as re
multiples = r'\|'
multi_class = filtered_chars[filtered_chars['justClass'].str.contains(multiples)]
multi_class[['justClass', 'alias']].sort_values('justClass')
```

```python
popular_race = filtered_chars[['alias', 'processedRace']].groupby('processedRace').count().sort_values('alias', ascending=False)
```

```python
casting = filtered_chars[['alias', 'castingStat']].groupby('castingStat').count().sort_values('alias', ascending=False)

str_cast = filtered_chars[filtered_chars['castingStat'] == 'Str']
dex_cast = filtered_chars[filtered_chars['castingStat'] == 'Dex']
con_cast = filtered_chars[filtered_chars['castingStat'] == 'Con']
weird_casting = pd.concat([str_cast, dex_cast, con_cast])


filtered_casting = filtered_chars[['alias', 'castingStat']].drop(weird_casting.index)
filtered_casting.groupby('castingStat').count().sort_values('alias', ascending=False)
filtered_chars[filtered_chars['castingStat'] == 'Int'].groupby('justClass').count().sort_values('alias', ascending=False).head(30)
```

```python
ac_max = filtered_chars[['alias', 'AC', 'justClass']].groupby(['AC', 'justClass']).count().sort_values('AC', ascending=False)

high_ac = filtered_chars[filtered_chars['AC'] > 25]

high_ac[['alias', 'race', 'processedRace', 'justClass', 'subclass', 'level', 'AC', 
         'Str', 'Dex', 'Con', 'Int', 'Wis', 'Cha',
         'choices']].sort_values('AC', ascending=False)
```

```python
common_level = filtered_chars[['alias', 'level']].groupby('level').count().sort_values('level')
```

```python
'''
graph by level (or for tier) - histogram
graph for each Abi score - box and whisker
graph by race - pie chart
graph by class - pie chart
graph by casting stat - pie chart
'''
```

```python
# Bar Chart of Characters by Tier of Play
tier_array = filtered_chars[['alias', 'tier']].groupby('tier').count().sort_values('tier')
tier_list = list(tier_array['alias'])
tier_labels = ["Levels 1-4", "Levels 5-10", "Levels 11-16", "Levels 17-20"]
tier_colors = ['green', 'blue', 'purple', 'red']

fig, ax = plt.subplots()

ax.bar(tier_array.index, tier_list, label=tier_labels, color=tier_colors)
ax.set_ylabel("Number of Characters")
ax.set_title("Number of Characters by Tiers of Play")
ax.legend(title="Tier Levels")
```

```python
# bar chart of levels

level_array = filtered_chars[['alias', 'level']].groupby('level').count().sort_values('level')
level_list = list(level_array['alias'])

fig, ax = plt.subplots()

ax.bar(level_array.index, level_list, label=level_array.index)
ax.set_ylabel("Number of Characters")
ax.set_xlabel("Character Level")
ax.set_xticks(level_array.index)
ax.set_title("Number of Characters by Level")
```

```python
# Ability Score Scatter Plots
ability_scores = ['Str', 'Dex', 'Con', 'Int', 'Wis', 'Cha']
abi_array = np.asarray(filtered_chars[ability_scores])
abi_color = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']
medianprops = dict(linestyle='-', linewidth=2, color='silver')
meanlineprops = dict(linestyle='--', linewidth=2.5, color='black')

fig, ax = plt.subplots()
ax.yaxis.grid(True, linestyle='-', which='major', color='lightgrey',
               alpha=0.5)
ax.set_ylabel("Score")
ax.set_xlabel("Ability")
abi_plot = ax.boxplot(abi_array, showmeans=True, meanline=True, meanprops=meanlineprops, 
                      medianprops=medianprops, patch_artist=True)
for patch, color in zip(abi_plot['boxes'], abi_color):
    patch.set_facecolor(color)
plt.xticks([1,2,3,4,5,6], ability_scores)
plt.title("Ability Score Spread")
fig.text(0.78, 0.015, '---', color='black', backgroundcolor='white',
         weight='bold', size='large')
fig.text(0.82, 0.015, ' Average Value', color='black', weight='roman',
         size='x-small')
plt.show()
```

```python
# race and class pie charts

# race data cleaning for chart
race_count = list(popular_race['alias'])
race_choice = list(popular_race.index)
r_min_value = 300

race_small_value = sum([x for x in race_count if x < r_min_value])
race_slices = [x for x in race_count if x >= r_min_value]
race_slice_labels = [choice for choice, count in zip(race_choice, race_count) if count > r_min_value]
race_slice_labels.append("Other")


# class data cleaning for chart
class_count = list(popular_class['alias'])
class_choice = list(popular_class.index)
c_min_value = 100

class_small_value = sum([x for x in class_count if x < c_min_value])
class_slices = [x for x in class_count if x >= c_min_value]
class_slice_labels = [choice for choice, count in zip(class_choice, class_count) if count > c_min_value]
class_slice_labels.append("Other")

# casting data cleaning for chart
casting_count = list(casting['alias'])
casting_choice = list(casting.index)
cast_min_value = 1800

cast_sm_value = sum([x for x in casting_count if x < cast_min_value])
cast_slices = [x for x in casting_count if x >= cast_min_value]
cast_slice_labels = [choice for choice, count in zip(casting_choice, casting_count) if count > cast_min_value]
cast_slice_labels.append("Other")

fig, axes = plt.subplots(1,3, figsize=(15, 15))
axes[0].pie(race_slices + [race_small_value], labels=race_slice_labels, autopct='%1.1f%%',
        pctdistance=.8, textprops={'size':'small'}, radius=1)
axes[0].set_title("Most Popular Race Choice")
axes[1].pie(class_slices + [class_small_value], labels=class_slice_labels, autopct='%1.1f%%', 
        pctdistance=.8, textprops={'size':'small'}, radius=1)
axes[1].set_title("Most Popular Class Choice")
axes[2].pie(cast_slices + [cast_sm_value], labels=cast_slice_labels, autopct='%1.1f%%', 
        pctdistance=.8, textprops={'size':'small'}, radius=1)
axes[2].set_title("Most Common Casting Stat")
plt.show()
```

After extensive analysis and cleaning of the character data, I found the following items:

Most common race: Human
Most common class: Fighter
Average ability score: 13
Highest average ability: Dex (14.5, vs 14.3 for Con)
Most common casting stat: Int
Most common tier of play: Tier 1 (Levels 1-4)
Most characters at Level 1

Some of my original hypotheses proved true (Humans, Fighters, and Dexterity are the most popular of their respective aspects of creating a D&D character), but I was also surprised by the overwhelming prevalence of Intelligence as a casting stat over Wisdom. The primary reason for that is, I believe, a quirk in the character creation process. Perhaps Int is a "default" casting stat, unless otherwise specified; the data are not clear on this. I discovered a few other items of interest during my analysis. First of all, there is a smaller swath of multiclass characters than I anticipated. Second, there is an enormous variety of races chosen for D&D characters, so much so that Human is only the largest minority at less than 21%. Third, there was a surprisingly high AC (armor class) stat for one character in particular, a 3rd-level Sorcerer (relatively low level), whose AC was 30, higher than most epic-level (17+) characters. I am not sure the reason for this, as the data do not give enough detail to explain it. At this point in my analysis, I ignored Armor Class as a useful metric to analyze characters.

My recommendations for further developing the game are as follows. First, the fact that Humans are the most common race, but only make up 21% of the total character pool suggests that there may be too many options for race choices. Having that many choices burdens both player and dungeon master with more options than they can reasonably deal with. I recommend simplifying the race options for character creation, to only a dozen at most. Too many more than that becomes needlessly challenging to balance and still provide a unique flair for a character.

Second, I recommend (specifically for this character creation platform) that the first-level character creation process be as user-friendly as possible. While the data suggest that this may be the case already, amplifying that aspect of the experience will result in increased traffic and usage of the platform. As long as the first-level character creation process is smooth, this guarantees that the platform will see lots of usage and interaction.

Finally, depending on the intended purpose of the platform, the mechanics of high-level character creation could be made even more user-friendly. As a D&D player, I know that creating a high-level character can be a tricky process, which is the most likely reason so few high-level characters are created using this platform. There are two options, given this data: one, craft the level-up process carefully in order to boost usage of the platform for higher-level characters; or two, cap character creation at a lower level (whether 5, 6, 7, or 10, to maintain most of the current users).
